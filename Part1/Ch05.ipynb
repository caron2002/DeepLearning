```python
# 역전파

# 곱셈계층

class MulLayer:
  def __init__(self):
    self.x = None
    self.y = None

  def forward(self, x, y):
    self.x = x
    self.y = y

    out = x * y
    return out

  def backward(self, dout):
    dx = dout * self.y
    dy = dout * self.x

    return dx, dy



# 사과쇼핑
# 사과 = 100, 사과 개수 = 2 tax = 1.1

apple = 100
apple_num = 2
tax = 1.1

mul_apple_layer = MulLayer()
mul_tax_layer = MulLayer()

apple_price = mul_apple_layer.forward(apple, apple_num)
price = mul_tax_layer.forward(apple_price, tax)

print(price)

# 역전파
dprice = 1
dapple_price, dtax = mul_tax_layer.backward(dprice)
dapple, dapple_num = mul_apple_layer.backward(dapple_price)

print(dapple, dapple_num, dtax)
```

```python
# 덧셈 계층

class AddLayer:
  def __init__(self):
    self.x = None
    self.y = None

  def forward(self, x, y):
    self.x = x
    self.y = y

    out = x + y
    return out

  def backward(self, dout):
    dx = 1 * dout
    dy = 1 * dout

    return dx, dy


# 사과 2개 귤 3개 구입 과정

apple = 100
apple_num = 2

orange = 150
orange_num = 3

mul_apple = MulLayer()
apple_price = mul_apple.forward(apple, apple_num)

mul_orange = MulLayer()
orange_price = mul_orange.forward(orange, orange_num)

add_total = AddLayer()
add_price = add_total.forward(apple_price, orange_price)

mul_tax = MulLayer()
price = mul_tax.forward(add_price, tax)

print(price)

#역전파
dprice = 1

dadd_price, dtax = mul_tax.backward(dprice)

dapple_price, dorange_price = add_total.backward(dadd_price)

dapple, dapple_num = mul_apple.backward(dapple_price)

dorange, dorange_num = mul_orange.backward(dorange_price)

print(f'dtax:\t\t{dtax}')
print(f'dapple:\t\t{dapple}')
print(f'dapple_num:\t{dapple_num}')
print(f'dorange:\t{dorange}')
print(f'dorange_num:\t{dorange_num}')

```

```python
# ReLU 순전파, 역전파
import numpy as np

class ReLU:
  def __init__(self):
    self.mask = None

  def forward(self, x):
    self.mask = (x <= 0)

    out = x.copy()
    out[self.mask] = 0

    return out

  def backward(self, dout):
    dout[self.mask] = 0
    dx = dout

    return dx



x = np.array([[1.0, -0.5], [-2.0, 3.0]])

relu = ReLU()
out = relu.forward(x)

print(out)

```

```python
# Sigmoid 순전파 역전파

class Sigmoid:
  def __init__(self):
    self.out = None

  def forward(self, x):
    out = 1 / (1 + np.exp(-x))

    self.out = out
    return out

  def backward(self, dout):
    dx = dout * self.out * (1.0 - self.out)

    return dx
```
